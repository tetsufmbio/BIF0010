##1.2 Transformando os campos do tipo *Object*

###1.2.1 Transformando o campo *Species* em One Hot Encoded 

```
from sklearn.preprocessing import LabelEncoder
# codifica o valor da classe como inteiros 
from keras.utils import np_utils
encoder = LabelEncoder()
encoder.fit(iris.Species)
encoded_Species = encoder.transform(iris.Species)
# converte inteiros para variáveis dummy (i.e. one hot encoded)
dummy_Species = np_utils.to_categorical(encoded_Species)
```

#2 Dividindo os Dados em Treinamento (train) e Teste (test) 

Podemos dividir nosso dataframe (**base_final**) em dois:

- Uma parte para treinar nosso modelo (80% da base): **train**
- Uma parte para fazer previsões e testar nosso modelo (20% da base): **test**


```
from sklearn.model_selection import train_test_split
#random_state = número. Isso é útil se você deseja resultados reproduzíveis

all_y = dummy_Species
all_X = base_final.drop(["Id"],axis=1)

train_X, test_X, train_y, test_y = train_test_split(all_X, all_y, test_size=0.20, random_state=42, stratify=all_y)
```

```
#O cálculo da acurácia e da matriz de confusão, não trabalha com dados no formato one hot encoded, 
#logo temos que converter para valores atômicos
test_y_class = np.argmax(test_y,axis=1)  #retorna o indíce do maior valor em cada linha(axis=1)
pred_y_class = np.argmax(y_pred,axis=1)

accuracy = accuracy_score(test_y_class, pred_y_class)
print("Acurácia do teste via accuracy_score:", accuracy)

print("*"*40)
cm = confusion_matrix(test_y_class, pred_y_class)
print(pd.DataFrame(cm, columns=['Iris-setosa', 'Iris-versicolor','Iris-virginica'], index=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']))
print("Acurácia do teste via confusion_matrix:", (cm[0,0]+cm[1,1]+cm[2,2])/np.sum(cm))
#print("*"*40)
```
